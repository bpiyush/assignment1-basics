### Unicode representation


* A single character is mapped to a unique code points (e.g., `a` is mapped to 97 (check by `ord(a)`)). In all, Unicode `U+10FFFF` (1,114,112) code points of which ~150K are in active use.
* We can store it in an encoding, like `UTF-32` as 8-bytes (8 x 4 = 32 bits). The first 4 bytes specify the byte order (Byte Order Mark or `BOM`)
```py
$ "😁".encode("utf-32")
> b'\xff\xfe\x00\x00\x01\xf6\x01\x00'
```

* In this case, `\xff\xfe\x00\x00` `(FF FE 00 00)` represent the BOM and `\x01\xf6\x01\x00` (by reversed endianness: `00 01 F6 01`). This can be converted to `0x0001F601` and in turn to `U+1F601`.
* This `U+1F601` is $16^{4} \times 1 + 16^{3} \times 15 + 16^{2} \times 6 + 16^{1} \times 0 + 16^{0} \times 1 = 65536 + 61440 + 1536 + 0 + 1 = 128513$. We can verify this as follows:
```py
$ ord("😁")
> 128513
```
* Much more commonly, you’ll see Unicode text encoded as either UTF-8 or UTF-16. These are both variable-length encodings, made up of 8-bit or 16-bit units, respectively. In these schemes, code points with smaller index values take up fewer bytes, which saves a lot of memory for typical texts. The trade-off is that processing UTF-8/16 texts is more programmatically involved, and likely slower.

Reference: [A Programmer’s Introduction to Unicode](https://www.reedbeta.com/blog/programmers-intro-to-unicode/)
